```python
import numpy as np
data = np.loadtxt(fname='../week-four/inflammation-01.csv', delimiter=',')
print(data)
```

```python
fig = plt.figure(figsize=(10.0, 3.0))

fig.suptitle('This is my super long title', va='bottom', fontsize=16)
sbplot1 = fig.add_subplot(1, 3, 1)
sbplot2 = fig.add_subplot(1, 3, 2)
sbplot3 = fig.add_subplot(1, 3, 3)

sbplot1.set_ylabel('average')
sbplot1.plot(np.mean(data, axis=0), '--')

sbplot2.set_xlabel('time (days)')
sbplot2.set_ylabel('max')
sbplot2.plot(np.max(data, axis=0), '-o')

sbplot3.set_ylabel('min')
sbplot3.plot(np.min(data, axis=0), '--*')

fig.tight_layout()
plt.show()
```

## Lists

```python
odds = [1, 3, 5, 7]
print ("odds are:", odds)

print('first element:', odds[0])

print('last element:', odds[3])
print('last element:', odds[-1])
print('last element:', odds[len(odds)-1])
print('second to last element:', odds[-2])
```

```python
for numbers in odds:
    print(numbers)

print("\n")

for numbers in range(0, len(odds)):
    print(odds[numbers])
```
