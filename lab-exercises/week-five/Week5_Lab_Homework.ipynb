## Part 1 (Figures)

````python
import numpy as np
data = np.loadtxt(fname='../week-four/inflammation-01.csv', delimiter=',')
print(data)
````

````python
import matplotlib.pyplot as plt
````

````python
#1
std_inflammation = np.std(data, axis = 0)
std_plot = plt.plot(std_inflammation)
plt.xlabel('time (days)')
plt.ylabel('standard deviation of inflammation score')
plt.title('patient standard deviation data')
plt.show()
````

````python
fig = plt.figure(figsize=(5.0, 10.0))

fig.suptitle('STDs, MAXs, & MINs \n', va='bottom', fontsize=12)
sbplot1 = fig.add_subplot(3, 1, 1)
sbplot2 = fig.add_subplot(3, 1, 2)
sbplot3 = fig.add_subplot(3, 1, 3)

sbplot1.set_ylabel('Standard Deviation')
sbplot1.plot(np.std(data, axis=0), '--')

sbplot2.set_ylabel('max')
sbplot2.plot(np.max(data, axis=0))

sbplot3.set_xlabel('time (days)')
sbplot3.set_ylabel('min')
sbplot3.plot(np.min(data, axis=0))

fig.tight_layout()
plt.show()
````
