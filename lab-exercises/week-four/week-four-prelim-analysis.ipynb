```python
# import numpy to be able to use numpy commands
import numpy
```

```python
# due to csv file containing floats, ints, and strings, it's best to import pandas then transfer data to numpy
# pandas also allowed me to make plots which allowed me to better visualize the data
import pandas as pd
tmp_data = pd.read_csv('Potential_Dataset.csv')
data = tmp_data.to_numpy()
```

```python
# show the numpy array
print(data)
```

```python
# confirm that it is an array
print(type(data))
```

```python
# just wanted to see what data type it showed up as
print(data.dtype)
```

```python
# shows rows vs columns
print(data.shape)
```

```python
# finding the middle value of columns and middle value of rows as well as middle value of data
data_shape = data.shape
middle_of_rows = int(data_shape[0]/2)
middle_of_cols = int(data_shape[1]/2)
print(middle_of_rows, middle_of_cols)

print('middle value of potential data:', data[middle_of_rows, middle_of_cols])
```

```python
# using pandas to see the first couple of columns (first 5 and last 5 entries)
df = pd.DataFrame(tmp_data, columns = ['No.', 'Location'])
df
```

```python
# table of first five sets of Foliar area vs latitude
df = pd.DataFrame(tmp_data,columns=['Foliar Area_mm2','Latitude'])
df.head()
```

```python
# plot of Foliar area vs latitude
df.plot(x ='Latitude', y='Foliar Area_mm2', kind = 'scatter')
```

```python
# table of first five sets of Foliar area vs longitude
df = pd.DataFrame(tmp_data,columns=['Foliar Area_mm2','Longitude'])
df.head()
```

```python
# plot of Foliar area vs longitude
df.plot(x ='Longitude', y='Foliar Area_mm2', kind = 'scatter')
```

```python
# table of first five sets of Foliar area vs Leaves
df = pd.DataFrame(tmp_data,columns=['Foliar Area_mm2','Leaf'])
df.head()
```

```python
# plot of Foliar area vs Leaves
df.plot(x ='Leaf', y='Foliar Area_mm2', kind = 'scatter')
```

```python
# seeing how often a species comes up
tmp_data['Species'].value_counts()
```

```python
# seeing which mean temps are common
tmp_data['Annual Mean Temperature, C'].value_counts()
```

```python
# plot of mean temp vs Leaves
df = pd.DataFrame(tmp_data,columns=['Annual Mean Temperature, C','Leaf'])
df.plot(x ='Leaf', y='Annual Mean Temperature, C', kind = 'scatter')
```

```python
# plot of Foliar area vs mean temp
df = pd.DataFrame(tmp_data,columns=['Annual Mean Temperature, C','Foliar Area_mm2'])
df.plot(x ='Foliar Area_mm2', y='Annual Mean Temperature, C', kind = 'scatter')
# appears to have a trend wher high foliar area can live in higher temps
```

```python
# seeing common precip values and least common precip values
tmp_data['Annual Precipitation (mm)'].value_counts()
```

```python
# plot of Foliar area vs precip
df = pd.DataFrame(tmp_data,columns=['Annual Precipitation (mm)','Foliar Area_mm2'])
df.plot(x ='Foliar Area_mm2', y='Annual Precipitation (mm)', kind = 'scatter')
# appears to have a trend wher high foliar area can live in high rain
```

```python
# plot of precip vs Leaves
df = pd.DataFrame(tmp_data,columns=['Annual Precipitation (mm)','Leaf'])
df.plot(x ='Leaf', y='Annual Precipitation (mm)', kind = 'scatter')
```
